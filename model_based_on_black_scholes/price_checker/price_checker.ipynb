```python
import csv
import pandas as pd
import matplotlib.pylab as plt
from datetime import datetime
import numpy as np
```

```python
path_to_file = '/Users/vladgusev/Desktop/merged_data.csv'
data = pd.read_csv(path_to_file, encoding='utf-8')
```

```python
final_data = {}
final_data['company'] = []
final_data['ticker'] = []
final_data['dividend release day'] = []
final_data['stock price day before'] = []
final_data['stock price on release day'] = []
final_data['stock price day after'] = []
final_data['dividend payment'] = []
final_data['put options day before'] = []
final_data['put options on release day'] = []
final_data['put options day after'] = []
```

```python
for index, row in data.iterrows:
    final_data['company'].append(data['Company'][index])
    final_data['company'].append(data['Ticker'][index])
    final_data['dividend release day'].append(data['DivDat1'][index])
    previous_div_payment = data['DivDat2'][index]
    zx = datetime.strptime(data['Last_DivDate'][index], "%y/%m/%d")
    st = 'Stock_day'
    i = 1
    while True:
        st += str(i)
        dt = datetime.strptime(data[st][index], "%y/%m/%d")
        if dt - zx == -1:
            vb = st + str(i - 1)
            stc = "Stock_on" + vb
            if data[stc][index] - data[st][index] > 0.1:
                final_data['stock price day before'].append('decreased')
            else if data[stc][index] - data[st][index] < 0.1:
                final_data['stock price day before'].append('increased')
            else:
                final_data['stock price day before'].append('not changed')
        elif dt - zx == 0:
            vb = st + str(i - 1)
            stc = "Stock_on" + vb
            if data[stc][index] - data[st][index] > 0.1:
                final_data['stock price on release day'].append('decreased')
            else if data[stc][index] - data[st][index] < 0.1:
                final_data['stock price on release day'].append('increased')
            else:
                final_data['stock price on release day'].append('not changed')
        elif dt - zx == 1:
            vb = st + str(i - 1)
            stc = "Stock_on" + vb
            if data[stc][index] - data[st][index] > 0.1:
                final_data['stock price day after'].append('decreased')
            else if data[stc][index] - data[st][index] < 0.1:
                final_data['stock price day after'].append('increased')
            else:
                final_data['stock price day after'].append('not changed')
            break;
    if (data['DivDat1'][index] < previous_div_payment):
        final_data['dividend payment'].append('decreased')
    else:
        final_data['dividend payment'].append('increased')
    
    i = 1
    while True:
        st += str(i)
        dt = datetime.strptime(data[st][index], "%y/%m/%d")
        if dt - zx == -1:
            vb = st + str(i - 1)
            stc = "Option_on" + vb
            if data[stc][index] - data[st][index] > 0.1:
                final_data['put options day before'].append('decreased')
            else if data[stc][index] - data[st][index] < 0.1:
                final_data['put options day before'].append('increased')
            else:
                final_data['stock price day after'].append('not changed')
        elif dt - zx == 0:
            vb = st + str(i - 1)
            stc = "Option_on" + vb
            if data[stc][index] - data[st][index] > 0.1:
                final_data['put options on release day'].append('decreased')
            else if data[stc][index] - data[st][index] < 0.1:
                final_data['put options on release day'].append('increased')
            else:
                final_data['stock price day after'].append('not changed')
        elif dt - zx == 1:
            vb = st + str(i - 1)
            stc = "Option_on" + vb
            if data[stc][index] - data[st][index] > 0.1:
                final_data['put options day after'].append('decreased')
            else if data[stc][index] - data[st][index] < 0.1:
                final_data['put options day after'].append('increased')
            else:
                final_data['put options day after'].append('not changed')
            break;
```

```python
df = pd.DataFrame(final_data)
```
